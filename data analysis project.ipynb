```python
#ANOVA(Analysis of variance) Test
from scipy import stats
df_anova=df1[["make","price"]]
df_test=df_anova.groupby(["make"],as_index=False)
anova_results=stats.f_oneway(df_test.get_group("honda")["price"],df_test.get_group("subaru")["price"])
print(anova_results)
```

```python
# here f value is <1 and p value is >0.5 so honda and subaru are similar
```

```python
anova_results1=stats.f_oneway(df_test.get_group("honda")["price"],df_test.get_group("jaguar")["price"])
print(anova_results1)
```

```python
# here f value is >1 and 400 and p value is >0.5 so honda and jaguar are variable with jaguar having high price
```
